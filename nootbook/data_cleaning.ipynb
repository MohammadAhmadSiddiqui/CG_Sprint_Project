```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Load Dataset

df = pd.read_csv(r"D:\New Sprint\dataset\diabetes_prediction_dataset.csv")

print("Dataset Shape:", df.shape)

df.head()
```

```python
# Check Dataset Information

df.info()
```

```python
# Check Missing Values

print(df.isnull().sum())
```

```python
# Remove Duplicate Rows

print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

```python
# Check Unique Values

print(df["gender"].unique())
print(df["smoking_history"].unique())
```

```python
# Remove Invalid Gender Category

df = df[df["gender"] != "Other"]

print(df["gender"].unique())
```

```python
# Remove BMI Outliers

df = df[(df["bmi"] > 10) & (df["bmi"] < 60)]

print("Shape After BMI Cleaning:", df.shape)
```

```python
# Remove Blood Glucose Outliers

df = df[
    (df["blood_glucose_level"] > 50) &
    (df["blood_glucose_level"] < 400)
]

print("Shape After Glucose Cleaning:", df.shape)
```

```python
# Final Dataset Information

df.info()
```

```python
# Save Cleaned Dataset

df.to_csv(
    r"D:\New Sprint\dataset\processed\cleaned_data.csv",
    index=False
)

print("Cleaned dataset saved successfully!")
```
