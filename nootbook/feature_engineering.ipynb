```python
# Import Libraries

import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Load Cleaned Dataset

df = pd.read_csv(r"D:\New Sprint\dataset\processed\cleaned_data.csv")

df.head()
```

```python
# Encode Gender Column

gender_encoder = LabelEncoder()

df["gender"] = gender_encoder.fit_transform(df["gender"])

print(df["gender"].head())
```

```python
# Encode Smoking History Column

smoking_encoder = LabelEncoder()

df["smoking_history"] = smoking_encoder.fit_transform(df["smoking_history"])

print(df["smoking_history"].head())
```

```python
# Create BMI Category Feature

df["bmi_category"] = pd.cut(
    df["bmi"],
    bins=[0, 18.5, 25, 30, 100],
    labels=[0, 1, 2, 3]
)

print(df["bmi_category"].head())
```

```python
# Convert BMI Category to Integer

df["bmi_category"] = df["bmi_category"].astype(int)
```

```python
# Create Glucose Risk Feature

df["glucose_risk"] = df["blood_glucose_level"].apply(
    lambda x: 1 if x > 140 else 0
)

print(df["glucose_risk"].head())
```

```python
# Check Final Dataset

df.head()
```

```python
# Save Engineered Dataset

df.to_csv(
    r"D:\New Sprint\dataset\processed\engineered_data.csv",
    index=False
)

print("Feature engineered dataset saved successfully!")
```
